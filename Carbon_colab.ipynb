```python
from sklearn.preprocessing import StandardScaler
# from pycaret.internal.preprocess import StandardScaler
df=pd.read_csv("dummys_2.0.csv")
df=df[df["source_type_nan"]!=1].drop(labels="source_type_nan",axis=1)[["qplume","red","green","blue","yellow"]]
X=df.drop(labels="qplume",axis=1)
y=df["qplume"]
X_train, X_test,y_train, y_test = train_test_split(X,y ,random_state=42,test_size=0.25, shuffle=True)
sc = StandardScaler()
sc.fit(X_train)
X_train_tf = sc.transform(X_train)
X_test_tf = sc.transform(X_test)
```

```python
pd.DataFrame(X_train_tf)
```

```python
train_transform=pd.DataFrame(np.hstack((X_train_tf,y_train[:,np.newaxis])),columns=["red","green","blue","yellow","qplumes"])
test_transform=pd.DataFrame(np.hstack((X_test_tf,y_test[:,np.newaxis])),columns=["red","green","blue","yellow","qplumes"])
transform=pd.DataFrame(np.vstack((train_transform,test_transform)),columns=["red","green","blue","yellow","qplumes"])
transform
```

```python
transform.to_csv("standardized_data.csv",index=False)
```
